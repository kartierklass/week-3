# NULL / Missing Values

```python
data[data['Rating'].isnull()]
```

```python
data.dropna(subset=['Rating'], inplace=True)
```

```python
data[data['Rating'].isnull()]
```

There are 2 columns **Age** and **Rating** that contain NULL values. Age that contain NULL values can be replaced by statistical parameter like mean, median , mode or can be replaced by some global constant, we can also replace it with its preceeding or suceeding value too. For Rating as there is only 1 values we can even drop this NULL value.

**Rating** column contains values that are negative we can adjust them to 0. for **Established** column we can set NULL values or to a global constant. For **Easy Apply** , by looking at data we can say -1  could indicate false value and so we can set it to False.

# Location Standardization

```python
data['Location']
```

```python
# Handling Special Characters
data.loc[data['Location'] == 'India,In', 'Location'] = "India"
data.loc[data['Location'] == 'Australia Aus', 'Location'] = "Australia"
data.loc[data['Location'] == 'New York,Ny', 'Location'] = "New York"
```
